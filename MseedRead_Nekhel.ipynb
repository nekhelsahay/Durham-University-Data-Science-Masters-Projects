```python
for i in range(1,54):
  st= ob.read("./Data/seeddata32/TEST"+str(i)+".mseed")
  st[0]
  tr=st[0]; #extract first trace from container "st"
  print(str(i),":")
  print("\n")
  print('##########################################')
  display(tr.stats) #show all header file information
  print('##########################################')
  print(' ')
  dt=tr.stats.delta; # extract time sampling
  print("delta or dt:",dt)
  dur=tr.stats.endtime-tr.stats.starttime # compute duration from end-start
  print("duration:",dur," s")
  data=tr.data # extract the data
  time=np.linspace(0,dur,len(data)) # create time vector
  len(data),len(time)
  print("\n \n \n \n")
```
